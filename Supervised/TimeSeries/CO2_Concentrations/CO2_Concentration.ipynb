# $\boldsymbol{CO_2}$ concentration forecast
In this notebook, we're going to do some time series forecasting in order to predict $CO_2$ levels in Mauna Loa, Hawi'i. The data is a time series describing monthly readings of $CO_2$ levels in ppm (parts per million) ranging from 1958 to 2018. The data has been gathered by the University of California's Scripps Institution of Oceanography, and can be found [here](http://scrippsco2.ucsd.edu/data/atmospheric_co2/primary_mlo_co2_record). As you may very well know, temperatures around the world are rising due to global warming, which is influenced heavily by greenhouse gas emissions such as $CO_2$ because of their high reflectivity in the IR.

Our goal in this report will be to examine these $CO_2$ levels and attempt to forecast them into the future. First, we'll do some exploratory data analysis (EDA) in order to get a feel for the dataset, then we'll be using some classical time series modeling techniques in order to try and predict future $CO_2$ levels.

```python
# First, let's download the dataset and get it into a pandas DataFrame
# You should examine the file visually in a text editor or excel first to get
#     idea of what's in there; there's a description of the data and some background
# The csv file has a header that ends at line 52, and the data itself doesn't start
#     until line 57

import pandas as pd
import numpy as np

raw_data = pd.read_csv('co2_data.csv',
                      skiprows=57,
                      names=['Yr', 'Mn', 'DateExcel', 'Date', 'CO2', 'SeasonallyAdjusted', 
                             'Fit', 'SeasonallyAdjustedFit', 'CO2Filled', 'SeasonallyAdjustedFilled'])
# We're only going to use the columns 1, 2, 4, and 5 (6+ are preprocessed versions of 5)
data = raw_data[['Yr', 'Mn', 'Date', 'CO2']]
```

```python
# Let's plot this data to get a feel for it
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2)
data_range = data['CO2']>0
x = data['Date'][data_range]
y = data['CO2'][data_range]
ax1.plot(x, y)
ax1.set_xlabel('Year')
ax1.set_ylabel('$CO_2$ [ppm]')
ax1.set_title('All the data!')

data_range = range(len(data)//2-25, len(data)//2+25)
x = data['Date'].iloc[data_range]
y = data['CO2'].iloc[data_range]
ax2.scatter(x, y)
ax2.set_xlabel('Year')
ax2.set_ylabel('$CO_2$ [ppm]')
ax2.set_title('Some subset of data')

plt.subplots_adjust(right=1.75)
plt.show()
```

As you can see from the plots above, the $CO_2$ levels have been increasing year-over-year pretty consistently. The trend appears to be approximately linear, but with a slight positive second derivative. As we zoom in, however, we can see that there is some definite oscillatory behavior with a period of approximately one year. We can do more than say some hand-wavey things about the trend; we can quantify these claims using a linear regression and Fourier analysis.

However, I neglected to mention earlier that there are some erroneous outliers in the data which were ignored when making the above plots:

```python
data[data['CO2']<0]
```

We have to deal with these data points somehow. One of the basic assumptions of most time series models is that the step between data points is constant, so just dropping these data points will be problematic for this assumption. Instead, we're going to interpolate between valid data points somehow. But before we do any manipulation of the data, we need to split it into train and test sets to avoid any data leakage. I'll save the last 15% of the data points for the test set, and we can use the first 85% for training.

```python
train_data = data.iloc[:int(0.75*len(data))].copy()
test_data = data.iloc[int(0.75*len(data)):].copy()
```

Since none of the erroneous data points cluster in groups any larger than 3, we'll use a sliding window of size 5 and replace these data points with the average within that window.

```python
for dataset in [train_data, test_data]:
    for t in range(len(dataset)):
        if dataset.iloc[t,3] < 0:
            tmin = max(0, t-2)
            tmax = min(len(dataset), t+3)
            avg_data = dataset.iloc[tmin:tmax].copy()
            avg_data = avg_data[avg_data['CO2']>0]
            dataset.iloc[t,3] = np.mean(avg_data['CO2'])
```

## Trend + periodicity model

Now that the train/test split and data imputation is taken care of, we can start modeling the data. As a first attempt, we'll use linear regression to find the overall trend curve:
$$
y = a_1 x + a_2 x^2 + b,
$$
subtract this out, then look at a spectral density plot by taking the Fourier transform to find the periodicity and model that.

```python
from sklearn.linear_model import LinearRegression
import scipy as sp

# Fit quadratic trend line
x = train_data['Date'].values
X = np.array([x, x**2]).T
y = train_data['CO2'].values
trend_model = LinearRegression()
trend_model.fit(X, y)
print("Trend coefficients: y = a1*x + a2*x^2 + b")
print("a1={:.3e}, a2={:.3e}, b={:.3e}".format(*trend_model.coef_, trend_model.intercept_))

# Subtract off trend line
train_data_sub = train_data.copy()
test_data_sub = test_data.copy()
train_data_sub['CO2'] = train_data['CO2'] - trend_model.predict(X)
x = test_data['Date'].values
X = np.array([x, x**2]).T
test_data_sub['CO2'] = test_data['CO2'] - trend_model.predict(X)

fig, (ax1, ax2) = plt.subplots(1, 2)

# Find peak(s) of power spectrum
ft = np.fft.rfft(train_data_sub['CO2'].values)
freq = np.fft.rfftfreq(len(train_data_sub), d=1/12)
peak_idx = np.argpartition(np.abs(ft),-3)[-3:] # Take 3 largest peaks
peak_freq = freq[peak_idx]
peak_amp = np.abs(ft[peak_idx])/len(train_data_sub)
phase = np.angle(ft[peak_idx])
ax1.plot(freq, np.abs(ft))
ax1.set_xlabel('Frequency [1/year]')
ax1.set_ylabel('PSD [ppm $CO_2$ * year]')
ax1.set_title('Power spectrum of trend-subtracted $CO_2$ concentration')

# Plot test residuals assuming top 3 peak frequencies are largest periodic components
x = test_data['Date']
X = np.array([x, x**2]).T
y = test_data['CO2'] - (trend_model.predict(X) + np.dot(np.cos(2*np.pi*np.outer(x, peak_freq) + phase), peak_amp))
err1 = np.sqrt(np.mean(y**2))
ax2.plot(x, y, label='trend+periodic')
y = test_data['CO2'] - (trend_model.predict(X))
err2 = np.sqrt(np.mean(y**2))
ax2.plot(x, y, label='trend')
ax2.legend()
ax2.set_xlabel('Year')
ax2.set_ylabel('Residuals [ppm $CO_2$]')
ax2.set_title('Test residuals: (data - model)')

plt.subplots_adjust(right=2)
plt.show()

print("RMS error (trend+periodic): {:.3f}".format(err1))
print("RMS error (trend): {:.3f}".format(err2))
```

So we can see that the data expresses a clear seasonality with a period of one year. By modeling the time series as simply the quadratic trend with an annual periodicity, we can get some fairly small residuals (compared with the overall magnitude of the time series, which is ~350) with an rms error of about 2-2.5, but we can do better. There is still clearly a periodic trend in the residuals - it's just not exactly sinusoidal. Now that the trend has been subtracted off, we can look at a seasonal subseries plot (in this case a boxplot).

```python
# Seasonal subseries plot
subseries = []
for g in train_data_sub.groupby('Mn'):
    subseries.append(g[1]['CO2'].values)
plt.boxplot(subseries)
plt.xticks(np.arange(1,13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

# Try to fit a sinusoid to it
x = np.linspace(1,13,100)
y = 2.5*np.cos(2*np.pi*x/12 + 1.3*np.pi)
plt.plot(x,y)
plt.show()
```

We can see that the $CO_2$ levels are clearly highest in the spring, and lowest during the fall. Apparently, this is because plants start absorbing carbon from the atmosphere during the spring/summer when they're most active, and release it during the fall/winter when they start going into "hibernation". Check out an explanation [here](https://scripps.ucsd.edu/programs/keelingcurve/2013/06/04/why-does-atmospheric-co2-peak-in-may/). You can also see that it isn't a perfectly sinusoidal curve, since the magnitude of the derivative is a bit higher during the summer, and levels out during the winter months.

## ARIMA Model

Since our previous model did not quite capture the non-sinusoidal periodicity of the data, we need something more sophisticated. One such model that is used quite a bit in time series forecasting is the ARIMA (AutoRegressive Integrated Moving Average) model. After differencing the data $d$ times, an $ARIMA(p,d,q)$ model is described by
$$
X_t = \delta + \phi_1 X_{t-1} + \phi_2 X_{t-2} + \cdots + \phi_p X_{t-p} \\
\qquad + A_t - \theta_1 A_{t-1} - \theta_2 A_{t-2} + \cdots + \theta_q A_{t-q}.
$$
where $\phi_i$ and $\theta_j$ are parameters of the model, $X_t$ are the data points at $t$, and $A_t$ are random noise terms (assumed to have a standard normal distribution). The parameters $p$, $d$, and $q$ determine the degree of the AR component, the number of times the data is differenced, and the degree of the MA component.

Since there is a yearly seasonal component, we should difference the data with an interval of 12 to remove it, and then check again for stationarity.

```python
# Difference data until it looks stationary
def diff(arr, interval=1, d=1):
    for _ in range(d):
        arr = (arr - np.roll(arr, interval))[interval:]
    return arr

fig, axes = plt.subplots(3, 2)

(ax1,ax2,ax3) = axes[:,0]
y = train_data_sub['CO2'].values
ax1.set_title("Undifferenced autocorrelation")
ax1.acorr(y, maxlags=50)
ax1.set_xlabel("Lag")
ax2.set_title("Undifferenced power spectrum")
ax2.plot(np.fft.rfftfreq(len(y), d=1/12), np.abs(np.fft.rfft(y)))
ax2.set_xlabel("1/year")
ax3.set_title("Undifferenced run plot")
ax3.plot(y)

(ax1,ax2,ax3) = axes[:,1]
y = diff(y, interval=12)
y_ = y
ax1.set_title("Diff. autocorr. w/ interval=12")
ax1.acorr(y, maxlags=200)
ax1.set_xlabel("Lag")
ax2.set_title("Diff. power spectrum w/ interval=12")
ax2.plot(np.fft.rfftfreq(len(y), d=1/12), np.abs(np.fft.rfft(y)))
ax2.set_xlabel("1/year")
ax3.set_title("Diff. run plot w/ interval=12")
ax3.plot(y)

"""
(ax1,ax2,ax3) = axes[:,2]
y = diff(y, interval=20)
ax1.set_title("2nd-diff. autocorr. w/ interval=52")
ax1.acorr(y, maxlags=200)
ax2.set_title("Diff. power spectrum w/ interval=12")
ax2.plot(np.fft.rfftfreq(len(y), d=1/12), np.abs(np.fft.rfft(y)))
ax3.set_title("2nd-diff. run plot w/ interval=52")
ax3.plot(y)
"""

plt.subplots_adjust(right=2, top=2, hspace=0.3)
plt.show()
```

Looks like a single differencing with an interval of 12 should be sufficient. And the autocorrelation plot suggests that there is a significant correlation at lags $\lesssim 15$. And let's just use $q=0$ for now for simplicity's sake.

```python
from statsmodels.tsa.arima_model import ARIMA

order = (15, 0, 0)
model = ARIMA(diff(train_data_sub['CO2'].values, d=1, interval=12), order=order)
model_fit = model.fit(disp=0)# Check out the residuals
plt.plot(model_fit.resid)
plt.title('ARIMA{} residuals'.format(order))
plt.show()
```

We need a way to undifference the results of the forecast. This can be done by taking the initial data point and the model for the differenced time series and inverting it as so:
$$\Delta_t^{(n)} = \Delta_{t+i}^{(n-1)} - \Delta_{t}^{(n-1)} \implies \Delta_{t+i}^{(n-1)} = \Delta_t^{(n-1)} + \Delta_t^{(n)}.$$
where $(n)$ denotes the degree of differencing (so that $\Delta^{(0)}_t = X_t$), and $i$ is the differencing interval. We need to work backwards from the model for $\Delta_t^{(n)}$ to $\Delta_t^{(n-1)}$, all the way to $\Delta_t^{(0)}$ by using the first $n$ elements of the original time series as initial data.

```python
def undiff(Dd, D0_init, d=1):
    """ Multiple undifferencing with interval of 1 """
    for n in reversed(range(d)):
        Dd_ = np.zeros(len(Dd)+1)
        Dd_[:d-n] = diff(D0_init, d=n)
        for t in range(1, len(Dd_)):
            Dd_[t] = Dd_[t-1] + Dd[t-1]
        Dd = Dd_
    return Dd

def undiff_(Dd, D0_init, interval=1):
    """ Single undifferencing with arbirary interval """
    Dd_ = np.zeros(len(Dd) + interval)
    Dd_[:len(D0_init)] = D0_init
    for t in range(interval, len(Dd_)):
        Dd_[t] = Dd_[t-interval] + Dd[t-interval]
    return Dd_

# Test the undifferencing; all lines should lie over each other if correct
# Unmodified test data:
plt.plot(test_data['CO2'].values)
# Test data differenced and undifferenced three times with interval of one
plt.plot(undiff(diff(test_data['CO2'].values, d=3), D0_init=test_data.iloc[0:3,3].values, d=3))
# Test data differenced and undifferenced with interval of 12
plt.plot(undiff_(diff(test_data['CO2'].values, interval=12), D0_init=test_data.iloc[:12,3].values, interval=12))
plt.title("Undifferencing test (only one curve should be visible)")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

Y = test_data_sub['CO2'].values
y = undiff_(model_fit.forecast(steps=len(Y)-12)[0], D0_init=train_data_sub.iloc[-12:,3], interval=12)
ax1.plot(Y, label="Test data")
ax1.plot(y, label="ARIMA forecast")
ax1.legend()
ax1.set_title("Forecast comparison")

res = y-Y
err = np.sqrt(np.mean(res**2))
ax2.plot(res)
ax2.set_title("Residuals: (data - model)")

plt.subplots_adjust(right=2)
plt.show()

print("RMS error: {:.3f}".format(err))
```

The error in this forecast is greatly improved over the previous trend+periodicity model, with an RMS of ~1.2, roughly a 40% improvement over the old RMS of ~2.0. We can see that we have captured the seasonality of the data pretty efficiently, as there is no apparent periodicity in the residuals. There are some notable deviations of the model from the data though, such as typically underpredicting the seasonal minima/maxima, and being unable to account for the permanent jumps in the $CO_2$ level near timesteps 120 and 150 (corresponding to 2013 and 2016, respectively), though I don't fault it for that; it appears that there are several such jumps throughout the training data as well, at irregular multi-year intervals, which can't be modeled unless we take into account hundreds of past data points.

## Neural Network Model

The ARIMA model was a significant improvement over the naive sinusoidal model, but there is potentially still room for improvement. ARIMA only takes into account the values of the time series at previous steps when calculating the current one; a more accurate model may also depend on other variables such as the month of the year, temperature data, etc.

If we can turn this problem into a typical supervised learning problem where we predict the next timestep based on a set of previous timesteps, then we can train a more powerful model than ARIMA (which is linear in the past timesteps) such as a neural network.
